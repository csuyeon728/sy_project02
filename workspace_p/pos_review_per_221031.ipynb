```python
import pandas as pd

prd = pd.read_csv("hou_prd_details.csv")
prd = prd.drop('Unnamed: 0', axis = 1)          

house = pd.read_csv("housewarming2.csv")
                
prd_details_fixed = pd.read_csv('prd_details_fixed.csv')
prd_style_match = pd.read_csv('prd_style_match.csv')
prd_details_fixed = prd_details_fixed.drop('Unnamed: 0', axis = 1)
```

```python
import pandas as pd

review_senti = pd.read_csv("review_senti.csv")
review_senti.head()
review_senti.info()
```

```python
review_candidates = pd.read_csv('review_candidates.csv')
review_candidates.head()
```

```python
pd.set_option('display.max_rows', None)
prd_posneg = pd.DataFrame(review_senti.groupby('상품url')['sentiment'].value_counts())
prd_posneg
```

```python
prd_url = []
for i in range(0, len(prd_posneg.index), 3):
    prd_url.append(prd_posneg.index[i][0])
    
print(prd_url)
```

```python
posneg_per = []
for i in range(0, 3425, 3):
    a = (prd_posneg.iloc[i,:][0] / (prd_posneg.iloc[i,:][0]+prd_posneg.iloc[i+1,:][0]+prd_posneg.iloc[i+2,:][0])) * 100
    #print(prd_posneg.iloc[i,:][0])
    #print(prd_posneg.iloc[i+1,:][0])
    #print(prd_posneg.iloc[i+2,:][0])
    posneg_per.append(a)
    
    if i > 3422:
        break
    
    i += 1

print(posneg_per)
```

```python
prd_posneg = dict(zip(prd_url, posneg_per))
prd_posneg
```

```python
# 상품 긍정 리뷰순
prd_posneg_df = pd.DataFrame(list(prd_posneg.items()), columns=['상품url', '긍정리뷰%'])
prd_posneg_df.sort_values(by='긍정리뷰%', ascending=False)
```

```python
prd_posneg_perdf = prd_posneg_df.merge(prd_details_fixed[['상품명','상품url']], how='left', on ='상품url')
prd_posneg_perdf.head()
```

```python
prd_posneg_perdf = prd_posneg_perdf.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False).sort_values(by='긍정리뷰%', ascending=False)
prd_posneg_perdf.info()
```

```python
prd_posneg_perdf = prd_posneg_perdf.merge(prd[['상품url','카테고리']], how='left', on='상품url')
```

```python
prd_posneg_perdf['카테고리'] = prd_posneg_perdf['카테고리'].apply(lambda x : x.split(',')[0].lstrip('[').strip("'"))
prd_posneg_perdf = prd_posneg_perdf.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False).sort_values(by='긍정리뷰%', ascending=False)
prd_posneg_perdf
```
