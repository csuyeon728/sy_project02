##  > 각 카테고리를 포함하는 상품갯수의 비율 ?

```python
style = ['내추럴', '러블리&로맨틱', '북유럽', '빈티지&레트로', '유니크&믹스매치', '클래식&앤틱', '한국&아시아', '프렌치&프로방스','모던', '미니멀&심플', '인더스트리얼']

cnt = 0
cmt = 0
for i in style:
    for j in range(0,109):
        if i in prd_cd.iloc[j,0]:
            # prd_cd[:][prd_cd['스타일'].apply(lambda x : i in x)]:
            cnt = cnt + prd_cd.iloc[j,1]
        
        else:
            pass
        
    print(i, cnt)
    cmt += cmt + cnt
print(cmt)
```

### > 각 단일 스타일별 포함하는 상품갯수 비율

```python
style = ['내추럴', '러블리&로맨틱', '북유럽', '빈티지&레트로', '유니크&믹스매치', '클래식&앤틱', '한국&아시아', '프렌치&프로방스','모던', '미니멀&심플', '인더스트리얼']

cnt = 0
cmt = 0
for i in style:
    for j in range(0,109):
        if i in prd_cd.iloc[j,0]:
            # prd_cd[:][prd_cd['스타일'].apply(lambda x : i in x)]:
            cnt = cnt + prd_cd.iloc[j,1]
        
        else:
            pass
        
    print(i, cnt, round(cnt/116499436 * 100, 4))
    cmt += cmt + cnt
print(cmt)
```
