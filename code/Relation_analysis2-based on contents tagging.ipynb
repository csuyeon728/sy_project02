```python
# 데이터 임포트
import pandas as pd

prd_relation = pd.read_csv('prd_relation.csv')
prd_relation.head()
```

```python
# photo_id 컬럼 인덱스로 변경.
prd_relation.set_index('photo_id',inplace = True)
prd_relation
```

```python
prd_relation.info()
```

```python
# 데이터값 문자형 →숫자형 변환 → bool타입으로 변환
prd_relation = prd_relation[:].apply(lambda x : x.astype('int'))
prd_relation = prd_relation[:].apply(lambda x : x.astype('bool'))
prd_relation
```

```python
# 연관분석에 필요한 패키지 로드
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder   # 연관분석을 위한 데이터 전처리 모듈
from mlxtend.frequent_patterns import apriori          # 연관분석 모듈
```

```python
apriori(prd_relation, min_support = 0.01, use_colnames=True)
```

```python
# 지지도 0.001, 신뢰도와 향상도 출력
a = apriori(prd_relation, min_support=0.001, use_colnames=True)

from mlxtend.frequent_patterns import association_rules

association_rules(a, metric='confidence', min_threshold= 0.001)
```

```python
# ln[108] 출력결과 데이터프레임으로 변경
apriori_ct = pd.DataFrame(association_rules(a, metric='confidence', min_threshold= 0.001))
apriori_ct
```

```python
apriori_ct.to_csv('total_apriori.csv')
```

```python
# antecedents, consequents 데이터타입 frozensets에서 list 로 변경 (엑셀로 내려받아서 데코/식물 찾기도 가능)
apriori_ct['antecedents'] = apriori_ct['antecedents'].apply(lambda x : list(x))
apriori_ct['consequents'] = apriori_ct['consequents'].apply(lambda x : list(x))
apriori_ct
```

```python
apriori_ct.to_csv('apriori_prd_comb.csv')
```
